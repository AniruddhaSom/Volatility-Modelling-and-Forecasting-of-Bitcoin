Now we will do rolling statistics (rolling mean and standard deviation) to check moving average and volatality over time.
Here I took 30 days of trading activity which shows the short time monthly trend

```python
#30days rolling mean of closing price
Bitcoin_dfnew['RollingMean_30'] = Bitcoin_dfnew['Close'].rolling(window=30).mean()

#30days rolling standard deviation (volatility)
Bitcoin_dfnew['RollingStd_30'] = Bitcoin_dfnew['Close'].rolling(window=30).std()

plt.figure(figsize=(10,5))
plt.plot(Bitcoin_dfnew['Close'], label= 'Closing Price')
plt.plot(Bitcoin_dfnew['RollingMean_30'], label= '30day Rolling Mean')
plt.legend
plt.show()
```

This chart(below) shows Bitcoin’s closing price along with its 30-day rolling standard deviation, which measures volatility. We see that periods of rapid price increase, such as the 2013 rally and the 2017 run-up, are accompanied by high volatility. Conversely, in stagnant periods, volatility declines. Economically, this demonstrates the risk–return tradeoff: Bitcoin’s sharp gains are matched with higher uncertainty.
1. Blue line (Price trend)
2. Orange line (Volatality- 30 days rolling std.)

```python
plt.figure(figsize=(10,5))
plt.plot(Bitcoin_dfnew['Close'], label= 'Closing Price')
plt.plot(Bitcoin_dfnew['RollingStd_30'], label= '30day Rolling Std')
plt.legend
plt.show()
```

Moving Average technique

Blue line: Actual Closing Price of Bitcoin. (below graph)
Orange line: 20-day SMA → smooths out the last 20 daily prices (short-term trend).
Green line: 50-day EMA → smooths out the last 50 days but gives more weight to recent data (longer-term trend).
both SMA and EMA smooth out noise and help identify trends.
TREND IDENTIFICATION - When the blue line (price) stays above both averages, it indicates an uptrend.
                       When the price falls below them, it signals a downtrend.
CROSSOVERS AS SIGNALS - When the 20-day SMA (orange) crosses above the 50-day EMA (green) → bullish signal (potential start of an uptrend).
                        When the 20-day SMA falls below the 50-day EMA → bearish signal (potential start of a downtrend).                       

```python
# 20-day Simple Moving Average
Bitcoin_dfnew['SMA_20'] = Bitcoin_dfnew['Close'].rolling(window=20).mean()

# 50-day Exponential Moving Average
Bitcoin_dfnew['EMA_50'] = Bitcoin_dfnew['Close'].ewm(span=50, adjust=False).mean()

# Plot
plt.figure(figsize=(8,4))
plt.plot(Bitcoin_dfnew['Close'], label='Closing Price')
plt.plot(Bitcoin_dfnew['SMA_20'], label='20-day SMA')
plt.plot(Bitcoin_dfnew['EMA_50'], label='50-day EMA')
plt.legend()
plt.show()
```
